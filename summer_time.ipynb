◎アメリカ、カナダ、メキシコ（一部除く）  
≪開始日≫3月の第2日曜日午前2時  
≪終了日≫11月第1日曜日午前2時  

開始日→午前1:59の次は時計の針を1時間進めて午前3時に。  
終了日→午前1:59の次は時計の針を1時間戻して午前1時に。  
また、開始日は1日が23時間、終了日は逆に25時間になります。

[Pythonのdatetimeで夏時間を扱う](http://nekoya.github.io/blog/2013/07/02/python-aware-datetime-dst/)

```
import pandas

def is_dst(datetime_string='2017/03/12 15:59:59'):
  datetime = pandas.Timestamp(datetime_string)
  # convert Tokyo time
  datetime = datetime.tz_localize('Asia/Tokyo')
  print(datetime)
  # convert US time
  datetime = datetime.tz_convert('US/Eastern')
  print(datetime)
  print(datetime.dst())
  print(datetime.dst().seconds)
  return(datetime.dst().seconds > 0)

if __name__ == '__main__':
  print(is_dst('2017/03/12 15:59:59'))
  print(is_dst('2017/03/12 16:00:00'))
  
```
